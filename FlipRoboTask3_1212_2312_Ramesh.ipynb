```python
# 1.Write a python program which searches all the product under a particular product from www.amazon.in. 
# The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search 
# for guitars.

# 2.In the above question, now scrape the following details of each product listed in first 3 pages of your 
# search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
# results then scrape all the products available under that product name. Details to be scraped are: "Brand 
# Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
# “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.


import selenium
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
import time
import pandas as pd
#from selenium import webdriver
from selenium.webdriver.chrome.service import Service
Webdriver_Path = Service("C:\\DataTrained\\FlipRobo\\chromedriver_win32\\chromedriver.exe")
driver = webdriver.Chrome(service=Webdriver_Path)
#driver = webdriver.Chrome(r"C:\DataTrained\FlipRobo\chromedriver_win32\chromedriver.exe")

# opening the URL
driver.get("https://www.amazon.in/")
#maximize the window size  
driver.maximize_window()  
#delete the cookies  
driver.delete_all_cookies()
product_search = input("enter product name to be searched ")
driver.find_element(By.ID, "twotabsearchtextbox").send_keys(product_search)
driver.find_element(By.ID, "nav-search-submit-button").click()
time.sleep(3)

ProductURL = []
Title=[]
price = []
Exp_Delivery = []
Return_Exchange = []
Availability = []
Brand =[]


#Scrap Product URL
start = 0
last = 3
for page in range(start,last):
    time.sleep(3)
    url = driver.find_elements(By.XPATH, "//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
    for i in url:
        ProductURL.append(i.get_attribute("href"))
               
    next_button =driver.find_element(By.XPATH, "//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_button.click()
    time.sleep(2)
    
#print(len(ProductURL))

for url in ProductURL[0:100]:
    time.sleep(3)
    driver.get(url)
    time.sleep(2)
     
    try:
        brand = driver.find_element(By.XPATH,"//*[@id='productOverview_feature_div']/div/table/tbody/tr[1]/td[2]")
        Brand.append(brand.text.strip())
    except NoSuchElementException:
        Brand.append('-')
        
    try:
        available = driver.find_element(By.XPATH,"(//div[@id='availability']//span)[1]")
        Availability.append(available.text.strip())
    except NoSuchElementException:
        Availability.append('NA')
        
    try:
        ReturnExch = driver.find_element(By.XPATH,"//*[@id='RETURNS_POLICY']/span/div[2]/a")
        Return_Exchange.append(ReturnExch.text.strip())
    except NoSuchElementException:
        Return_Exchange.append('-')
   
    try:
        title = driver.find_element(By.XPATH, "//div[@class='a-section a-spacing-none']//span[@id='productTitle']")
        Title.append(title.text.strip())
    except NoSuchElementException:
        Title.append('-')
        
    try:
        price_scrap = driver.find_element(By.XPATH, "(//div[@class='a-section a-spacing-none aok-align-center']//span[@class='a-price-whole'])[1]")
        price.append(price_scrap.text.strip())
    except NoSuchElementException:
        price.append('-')
    
    try:
        expDelivery = driver.find_element(By.XPATH, "(//div[@id='mir-layout-DELIVERY_BLOCK'])[1]")
        Exp_Delivery.append(expDelivery.text.strip())
    except NoSuchElementException:
        Exp_Delivery.append('-')
    

#close the browser  
driver.close()  

#,'ProductURL':ProductURL
Amazon_list = {'Brand':Brand,'Name':Title,'price':price,'Return_Exchange':Return_Exchange,'Exp_Delivery':Exp_Delivery,
               'Availability':Availability,'ProductURL':ProductURL[0:100]}
df = pd.DataFrame(Amazon_list,columns =['Brand','Name','price','Return_Exchange','Exp_Delivery','Availability','ProductURL'])
display(df)
df.to_csv('AmazonTask3file'+str(product_search)+'.csv')
```

```python
# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
# “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, 
# overall reviews, privates from price, dorms from price, facilities and property description.
import selenium
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time
import pandas as pd
#from selenium import webdriver
from selenium.webdriver.chrome.service import Service
Webdriver_Path = Service("C:\\DataTrained\\FlipRobo\\chromedriver_win32\\chromedriver.exe")
driver = webdriver.Chrome(service=Webdriver_Path)
#driver = webdriver.Chrome(r"C:\DataTrained\FlipRobo\chromedriver_win32\chromedriver.exe")

# opening the URL
driver.get("https://www.hostelworld.com/london/")
#maximize the window size  
driver.maximize_window()  
#delete the cookies  
driver.delete_all_cookies()
time.sleep(2)
driver.implicitly_wait(20) # seconds


hostelname =[]
distance=[]
ratings = []
total_reviews = []
overall_reviews = []
privatesfrom_price = []
dormsfrom_price = []
facilities = []
property_description = []
# driver.execute_script("window.scrollBy(0,1000)")

#london_tag = driver.find_element(By.XPATH,'//*[@id="location-text-input-field"]')
#london_tag.send_keys("London, England")
#london_tag.send_keys(Keys.ENTER)
# london_tag.click()
#WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//input[@class="search-input"]'))).send_keys("London")
# time.sleep(4)

driver.execute_script("window.scrollBy(0,500)")

driver.find_element(By.XPATH,'//input[@class="location-input"]').click()
driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div[2]/div[1]/div/div/div[1]/div/label/input').send_keys("London")
driver.find_element(By.XPATH,'//input[@class="location-input"]').send_keys(Keys.ENTER)
time.sleep(2)
driver.execute_script("window.scrollBy(0,500)")
time.sleep(2)    
WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@class="search-button"]'))).click()
time.sleep(2)

start =0
end =2
for page in range(start,end):
    hostelname_tags = driver.find_elements(By.XPATH, "//h2[@class='title title-6']/a" or '//a[@class="property-card-container horizontal"]//div[@class="property-name"]')
    for i in hostelname_tags:
        hostelname.append(i.text.strip())
    #print(hostelname)   
    dist_tags = driver.find_elements(By.XPATH, "//div[@class='subtitle body-3']//span[@class='description']" or "//div[@class='property-distance']//span[@class='distance-description']" )
    for i in dist_tags:
        distance.append(i.text.strip())

    #print(distance) 
    rating_tags = driver.find_elements(By.XPATH, "//div[@class='bottom-rating']//div[@class='rating rating-summary-container big']//span" or "//a[@class='property-card-container horizontal']//div[@class='property-rating']//div[@class='score-row']")
    for i in rating_tags:
        try:
            ratings.append(i.text.strip())
        except NoSuchElementException:
            ratings.append('No Rating')

    #print(ratings)

    totreviews_tags = driver.find_elements(By.XPATH, "//div[@class='summary orange'  or @class='summary gray']//div[@class='reviews']" or
                                          "//a[@class='property-card-container horizontal']//div[@class='review']//span[@class='left-margin']")

    for i in totreviews_tags:
        try:
            total_reviews.append(i.text.strip())
        except NoSuchElementException:
               total_reviews.append('-')
        
    #print(total_reviews)  


    privateprice_tags = driver.find_elements(By.XPATH, "//div[@class='price-col'][1]" or "(//a[@class='property-card-container horizontal']//div[@class='property-accommodations']//div[@class='accommodation-price single-row'])[1]")
    for i in privateprice_tags:
        try:
            privatesfrom_price.append(i.text.strip())
        except NoSuchElementException:
            privatesfrom_price.append('-')
    #print( privatesfrom_price) 
    
    dormsprice_tags = driver.find_elements(By.XPATH, "//div[@class='price-col'][2]" or"(//a[@class='property-card-container horizontal']//div[@class='property-accommodations']//div[@class='accommodation-price single-row'])[1]")
    for i in dormsprice_tags:
        try:
             dormsfrom_price.append(i.text.strip())
        except NoSuchElementException:
            dormsfrom_price.append('-')
    #print( dormsfrom_price) 
    
    facilities_tags = driver.find_elements(By.XPATH, "//div[@class='facilities-label facilities']" or "//div[@class='property-badges']")
    for i in facilities_tags:
        try:
            facilities.append(i.text.strip())
        except NoSuchElementException:
            facilities.append('-')
    #print( facilities)

   
    driver.execute_script("window.scrollBy(0,9000)")
    time.sleep(4)
    driver.find_element(By.XPATH, "//i[@class='core-icon icon-core-chevron-right']" or "//button[@class='pill-content page-nav nav-right icon-only']").click()
    

#close the browser  
driver.close()
#,'property_description':property_description
hostel_list = {'hostelname':hostelname,'distance':distance,'ratings':ratings,'total_reviews':total_reviews,
                 'privatesfrom_price':privatesfrom_price,'dormsfrom_price':dormsfrom_price,
                 'facilities':facilities}
df = pd.DataFrame(hostel_list,columns =['hostelname','distance','ratings','total_reviews','privatesfrom_price',
                                        'dormsfrom_price','facilities'])
df
```
